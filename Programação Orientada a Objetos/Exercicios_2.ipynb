```python
class Bola:
    '''
    Cria uma Bola
    
    Parâmetros
    ----------
    cor : str
        Cor da bola
    raio : float
        Raio da bola       
    '''
    def __init__(self, cor, raio):
        self.cor = cor
        self.raio = raio
        
    def imprime_cor(self):
        '''
        Imprime a cor da bola de self.
        
        Retorno
        -------
        out : str
            Imprime a cor da bola.        
        '''
        print(f"A cor da bola é: {self.cor}.")
        
    def calcula_area(self):
        '''
        Calcula a área da bola de self.
        
        Retorno
        -------
        out : float
            Retorna a área da bola.
        '''        
        area = (4 * 3.14 * (self.raio * self.raio))
        return area
    
    def calcula_volume(self):
        '''
        Calcula o volume da bola de self.
        
        Retorno
        -------
        out : float
            Retorna o volume da bola.
        '''
        volume = ((4/3) * 3.14 * (self.raio * self.raio * self.raio))
        return volume
```

```python
class Agenda:
    '''
    Cria uma agente com contatos (nome, telefone e e-mail).
    
    Parâmetros
    proprietario : str
        Nome do proprietário da agenda.
    '''
    def __init__(self, proprietario):
        self.proprietario = proprietario
        self.lista_contatos = []
        
    def __repr__(self):
        '''
        Representação do proprietário.
        
        Retorno
        -------
        out: str
            Retorna o proprietário da agenda.
        '''
        return f'Agenda de Propriedade de {self.proprietario}.'
    
    def adicionar_contato(self, nome, telefone, email):
        '''
        Adiciona um contato, utilizando um dicionário, na lista de contatos.
        O critério para identificar um contato é o e-mail, por ser único,
        pois podemos ter pessoas diferentes com o mesmo nome (pessoas homônimas).
        
        Parâmetros
        ----------
        nome : str
            Nome do contato
        telefone : str
            Telefone do contato. Deve estar no seguinte formato: "(DDD) XXXXX-XXXX"
        email : str
            E-mail do contato.
            
        Retorno
        -------
        out : bool
            Retorna True, se o contato foi adicionado.
            Retorna False, se o contato já existe na lista de contatos.
        '''
        info_contato = {
            "Nome": nome,
            "Telefone": telefone,
            "E-mail": email            
        }
        
        # Verificando se a lista está vazia:
        if self.lista_contatos == []:
            # Adiciona-se o contato e retorna True
            self.lista_contatos.append(info_contato)
            return True
        else:
            # Verificando se o e-mail já existe e retorna False
            for contato in self.lista_contatos:
                if info_contato["E-mail"] == contato["E-mail"]:
                    return False
            
            # Caso o e-mail não exista, adicionar e retorna True
            self.lista_contatos.append(info_contato)
            return True
        
    def remover_contato(self, email):
        '''
        Remove o contato a partir do E-mail, que é único.
        
        Paramêtro
        ---------
        email : str
            E-mail do contato.
            
        Retorno
        -------
        out : bool
            Retorna True, caso o contato tenha sido removido.
            Retorna False se o contato não existe.
            Retorna 0, se a lista de contatos está vazia.        
        '''
        
        # Verificando se a lista está vazia. Caso esteja, retorna 0
        if self.lista_contatos == []:
            return 0
        else:
            # Verificando se existe o contato com o e-mail, remove e retorna True
            for contato in self.lista_contatos:
                if email == contato["E-mail"]:
                    indice = self.lista_contatos.index(contato)
                    self.lista_contatos.pop(indice)
                    return True
                
            # Caso não exista o contato, retorna False
            return False
        
    def buscar_contato(self, pesquisar):
        '''
        Busca os contatos que possuem o nome digitado (ou parte desse nome).
        
        Parâmetro
        ---------
        pesquisar : str
            Nome (ou parte desse nome digitado), do contato que se pesquisa (procura/busca).
        
        Retorno
        -------
        out : list
            Retorna a lista de nomes com os contatos pesquisados.
        
        '''
        # Lista com os nomes
        lista_nomes = []
        # Adiciona na lista de nomes os contatos com o pesquisar
        for contato in self.lista_contatos:
            if pesquisar in contato["Nome"]:
                lista_nomes.append(contato)
        return lista_nomes
        
    def listar_contatos(self):
        '''
        Mostra os contatos adicionados na sua lista de contatos.
        
        Retorno
        -------
        out : list
            Retorna a lista com todos os contatos
        '''
        return self.lista_contatos
```

```python
class Fracao:
    '''
    Cria uma classe fração, que permite realizar operações matemáticas.
    
    Parâmetros
    ----------
    numerador : float
        Valor do numerador
    denominador : float
        Valor do denominador. Lembre-se que não pode ser zero.
    '''
    def __init__(self, numerador, denominador):
        if(denominador == 0):
            raise ValueError('Denominador não pode ser 0.')
            
        self.numerador = numerador
        self.denominador = denominador
        
    def valor(self):
        '''
        Realiza o cálculo da fração entre numerador e o denominador.
        
        Retorno
        -------
        out : float
            Retorna o valor do cálculo da divisão entre numerador e denominador.
        '''
        return self.numerador/self.denominador
        
    def __repr__(self):
        '''
        Forma de se representar a fração.
        '''
        return f'{self.numerador}/{self.denominador}'
    
    # Agora, temos formas ágeis de se realizar cálculos e implementar comparações:
    
    def __add__(self, other):
        num = self.numerador*other.denominador +\
        self.denominador*other.numerador
        
        den = self.denominador*other.denominador
        
        return Fracao(num, den)
    
    def __sub__(self, other):
        num = self.numerador*other.denominador -\
        self.denominador*other.numerador
        
        den = self.denominador*other.denominador
        
        return Fracao(num, den)
    
    def __mul__(self, other):
        num = self.numerador * other.numerador
        den = self.denominador * other.denominador
        
        return Fracao(num, den)
    
    def __truediv__(self, other):
        num = self.numerador*other.denominador
        den = self.denominador*other.numerador
        
        return Fracao(num, den)
    
    def __eq__(self, other):
        return self.valor() == other.valor()
    
    def __lt__(self, other):
        return self.valor() < other.valor()
    
    def __le__(self, other):
        return self.valor() <= other.valor()
    
    def __gt__(self, other):
        return self.valor() > other.valor()  
    
    def __ge__(self, other):
        return self.valor() >= other.valor()
```

```python
import datetime
class Data:
    '''
    Cria uma data.
    
    Parâmetros
    ----------
    dia : int
        Dia da data.
    mes : int
        Mês da data.
    ano : int
        Ano da data
    '''
    def __init__(self, dia, mes, ano):
        self.dia = dia
        self.mes = mes
        self.ano = ano
        self.periodo = datetime.datetime(self.ano, self.mes, self.dia)
    
    def __repr__(self):
        '''
        Formata a data.
        
        Retorno
        -------
        out : str
            Formata a data para o formato "dia/mês/ano": dd/mm/aaaa .
        '''
        return f'A data é {self.dia:02d}/{self.mes:02d}/{self.ano:04d}.'
    
    def comparando_datas_igual(self, other):
        '''
        Compara se a data de self é igual a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é igual a data de other.
            Retorna False se a data de self não é igual a data de other.
        '''
        if self.periodo == other.periodo:
            return True
        return False
        
    def comparando_datas_maior(self, other):   
        '''
        Compara se a data de self é maior que a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é maior que a data de other.
            Retorna False se a data de self não é maior que a data de other.
        '''
        if self.periodo > other.periodo:
            return True
        return False

        
    def comparando_datas_menor(self, other):    
        '''
        Compara se a data de self é menor que a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é menor que a data de other.
            Retorna False se a data de self não é menor que a data de other.
        '''
        if self.periodo < other.periodo:
            return True
        return False
        
    def comparando_datas_menor_maior(self, other):
        '''
        Compara se a data de self é menor/maior que a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é menor/maior que a data de other.
            Retorna False se a data de self não é menor/maior que a data de other.
        '''
        if self.periodo <= other.periodo:
            return True   
        return False
        
    def comparando_datas_maior_menor(self, other):  
        '''
        Compara se a data de self é maior/menor que a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é maior/menor que a data de other.
            Retorna False se a data de self não é maior/menor que a data de other.
        '''        
        if self.periodo >= other.periodo:
            return True
        return False
        
    def comparando_datas_diferente(self, other):
        '''
        Compara se a data de self é diferente que a data de other.
        
        Retorno
        -------
        out : bool
            Retorna True se a data de self é diferente que a data de other.
            Retorna False se a data de self não é diferente que a data de other.
        '''
        if self.periodo != other.periodo:
            return True
        return False
```
