# Avaliação módulo 1 - Degree Data Science

Na empresa em que você trabalha foi solicitada a implementação de um sistema de cadastro dos fornecedores. As informações que devem ser cadastradas são:

- Código do fornecedor;
- Nome do fornecedor;
- Telefone do fornecedor;
- Email do fornecedor.

Para isso, você deverá criar um script para armazenar essas informações em alguma estrutura de dados.

Também como parte do sistema de cadastro, é necessário criar uma "interface de navegação" pelo sistema, que tenha as seguintes funcionalidades:

- Permite a exibição (usando o `print`) de todas as informaçoes de contatos específicos, a partir da busca pelo código do fornecedor;
- Permite a remoção e adição de fornecedores à estrutura de cadastro.

Requisitos:

- Requisito #1 - Não existe um limite para fornecedores cadastrados, portanto você deverá permitir que o usuário cadastre a quantidade que desejar.
- Requisito #2 - Crie pelo menos uma função em seu script. Você deverá escolher pelo menos uma tarefa do seu script e implementar em uma função para chamá-la dentro da sua lógica de cadastro.
- Requisito #3 - Crie pelo menos uma verificação em seu código, por exemplo: Se o dicionário estiver vazio, não permita a opção remover.

Tópicos avaliados:

- Variáveis;
- Tipos de dados;
- Operadores aritméticos;
- I/O de dados;
- Expressões lógicas e condicionais;
- Laços de repetição;
- Listas e dicionários;
- Funções.


```python
def dados_cliente(a,b,d):
    idx_cod = d[a].index(b)
    return [v[idx_cod] for v in d.values()]


def cadastro(d):
    while True:
        if len(d['codigos']) == 0:
            operacao = (input
                    ('Cadastro vazio. Digite "A"=adição, "N"=saida :')).lower().strip()
            if operacao == "n" or operacao == 'a':
                break   
        else:   
            operacao = (input
                    ('Digite tipo de operaçaõ. "B"=busca, "R"=remoção, "A"=adição, "N"=saida : ')).lower().strip()
            if operacao == "b" or operacao == 'r' or operacao == 'a' or operacao == 'n' :
                break

                
    if operacao == 'b' or operacao == 'r':
        cod = input('Digite Código Cliente a ser procurado : ').strip()
        
        count = 0   
        while cod not in d['codigos']:
            cod = input("Código NÃO encontrado. Digite código novamente,:").strip()
            count += 1
            if count > 10:
                break

        if cod in d['codigos'] and operacao == 'b':
        
            cliente = dados_cliente('codigos', cod, d)

            print(cliente)        

        elif cod in d['codigos'] and operacao == 'r' :
        
            cliente = dados_cliente('codigos', cod, d)
            while True:
                print(f'Tem certeza de remover {cliente}')
                check = input('"S"=sim, "N"=saida :').strip().lower()
                if check == 's' or check == 'n':
                    break
                    
            if check == 's':
                idx_cod = d["codigos"].index(cod)
                for k in d:
                    d[k].pop(idx_cod)
                print(f'{cliente} removido com sucesso:',)
     
    
    if operacao == 'a':
        while True:
            cod = input('Digite código :').strip()
            if cod not in d['codigos']:
                break
            else:
                print('Código já em uso. Escolha outro código :')
      
        nome = input('Digite nome :').strip()
    
        tel = ''.join(filter(str.isnumeric,input('Digite telefone :').strip()))
        while len(list(tel)) < 9:
            tel = ''.join(filter(str.isnumeric,input('Digite telefone corretamente :').strip()))
                      
        email = input('Digite email :').strip()
        while '@' not in email or '.' not in email:
            email = input('Digite email corretamente :').strip()
    
        lista = [cod, nome, tel, email]
    
        for i,c in enumerate(d):
            d[c].append(lista[i])
    
        cliente = dados_cliente('codigos', cod, d)
        print(f'{cliente} cadastrado com sucesso:')
        

cadastro_cliente = {"codigos": ['0001' , '0002', '0003'],
               "nomes" : ['Alex', 'Leonardo', 'Luciana'], 
               "telefones": ["992651089", '997862354', '992567210'],
               "emails" : ['alex@gmail.com', 'leonardo@gmail.com', 'luci@gmail.com']}
while True:
    acesso_cadastro = input('Deseja acessar o cadastro cliente? "S"= sim, "N"=não :').strip().lower()
    if acesso_cadastro == 'n':
        break
    elif acesso_cadastro == 's':
        cadastro(cadastro_cliente)

cadastro_cliente        
```
